```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dados_df = pd.read_csv("Cars dataset.csv")
```

```python
dados_df
```

```python
dados_df.isnull().sum()
```

```python
# Analisando a distribuição de preços de venda
plt.figure(figsize=(10, 6))
sns.histplot(dados_df['selling_price'], bins=20, kde=True)
plt.title('Distribuição dos Preços de Venda')
plt.xlabel('Preço de Venda')
plt.ylabel('Frequência')
plt.show()
```

```python
# Analisando a relação entre Km rodados e preço de venda
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dados_df, x='km_driven', y='selling_price', hue='fuel')
plt.title('Relação entre Km Rodados e Preço de Venda')
plt.xlabel('Km Rodados')
plt.ylabel('Preço de Venda')
plt.show()
```

```python
# Analisando a média do preço de venda por tipo de combustível
fuel_price_mean = dados_df.groupby('fuel')['selling_price'].mean().sort_values(ascending=False)
fuel_price_mean.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Preço Médio de Venda por Tipo de Combustível')
plt.xlabel('Tipo de Combustível')
plt.ylabel('Preço Médio de Venda')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analisando a média do preço de venda por marca
brand_price_mean = dados_df.groupby('brand')['selling_price'].mean().sort_values(ascending=False)
brand_price_mean.plot(kind='bar', figsize=(12, 8), color='lightgreen')
plt.title('Preço Médio de Venda por Marca')
plt.xlabel('Marca')
plt.ylabel('Preço Médio de Venda')
plt.xticks(rotation=90)
plt.show()
```

```python
# Análise de Distribuição do Número de Proprietários por Tipo de Combustível
plt.figure(figsize=(10, 6))
sns.countplot(data=dados_df, x='fuel', hue='owner', palette='Set1')
plt.title('Distribuição de Proprietários por Tipo de Combustível')
plt.xlabel('Tipo de Combustível')
plt.ylabel('Número de Veículos')
plt.legend(title='Proprietário')
plt.show()
```

```python
brand_count = dados_df['brand'].value_counts().sort_values(ascending=False)
brand_count.plot(kind='bar', figsize=(12, 8), color='skyblue')
plt.title('Número de Veículos Vendidos por Marca')
plt.xlabel('Marca')
plt.ylabel('Número de Veículos Vendidos')
plt.xticks(rotation=90)
plt.show()
```

```python
# Comparando Preço Médio por Marca e Tipo de Combustível
brand_fuel_price_mean = dados_df.groupby(['brand', 'fuel'])['selling_price'].mean().unstack().sort_values(by='Diesel', ascending=False)
brand_fuel_price_mean.plot(kind='bar', figsize=(12, 8), colormap='viridis')
plt.title('Preço Médio por Marca e Tipo de Combustível')
plt.xlabel('Marca')
plt.ylabel('Preço Médio de Venda')
plt.xticks(rotation=90)
plt.legend(title='Combustível')
plt.show()
```

```python
# Análise de Tendências de Preço com Base no Número de Donos
dados_df['owner_type'] = dados_df['owner'].map({'First Owner': 1, 'Second Owner': 2, 'Third Owner': 3})
plt.figure(figsize=(10, 6))
sns.lineplot(data=dados_df, x='owner_type', y='selling_price', hue='fuel', marker='o')
plt.title('Tendência de Preço com Base no Tipo de Proprietário')
plt.xlabel('Tipo de Proprietário')
plt.ylabel('Preço de Venda')
plt.xticks([1, 2, 3], ['First Owner', 'Second Owner', 'Third Owner'])
plt.show()
```

```python
# Análise de Correlação Não Linear - Preço de Venda e Km Rodados
plt.figure(figsize=(10, 6))
sns.regplot(data=dados_df, x='km_driven', y='selling_price', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Correlação Não Linear entre Km Rodados e Preço de Venda')
plt.xlabel('Km Rodados')
plt.ylabel('Preço de Venda')
plt.show()
```
